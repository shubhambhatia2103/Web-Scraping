# Web Scrapping

Importing basics libraries that are used for web scrapping.

*   Requests - A Python library used to send an HTTP request to a website and store the response object within a variable.
*   BeautifulSoup - A Python library used to extract the data from an HTML or XML document.
*   Pandas - A Python library used for Data Analysis. Which will be used to create a dataframe in the following project.

```python
from bs4 import BeautifulSoup
import requests
```

# Request code.

This step is the main responsibility of the requests library. The `“requests.get()”` method sends an HTTP request to the specified URL and stores its response into a variable.

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"
page = requests.get(url)
```

# Use BeautifulSoup to store the HTML code.

Another feature of the requests library is the “requests.content” which returns all the data of an HTTP response. The BeautifulSoup stores this content (data) in an HTML format in the variable s.

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = "wikitable sortable")
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
world_title = table.find_all('th')
```

```python
world_title
```

```python
world_title_table = [title.text.strip() for title in world_title]
```

```python
print(world_title_table)
```

# Pandas DataFrame


```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_title_table)
```
